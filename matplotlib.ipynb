```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel("y-axes")
plt.title("2D-graph",loc='center')
plt.savefig("mat.jpg")
```

```python
y=x*x


plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('x-axis')
plt.ylabel("y-axes")
plt.title("2D-graph",loc='center')
```

```python
plt.subplot(3,1,1)
plt.plot(x,y,'r')
plt.subplot(3,1,2)
plt.plot(x,y,'g*-')
plt.subplot(3,1,3)
plt.plot(y,x,'bo-')
```

```python
x=np.arange(0,5 * np.pi ,0.1)
y=np.sin(x)
plt.plot(x,y,'r*')
plt.title("sinewaveform")
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'b*')
plt.title("sin graph")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'go')
plt.title("cos graph")
```

```python
data=[np.random.normal(2,std,200) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.xlabel('range')
plt.ylabel('orange')
plt.title("pickel")
```
